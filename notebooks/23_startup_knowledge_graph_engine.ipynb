```python
import pandas as pd
import networkx as nx

import joblib
import json
```

```python
df = pd.read_csv(
    "../datasets/cleaned/startup_info.csv",
    low_memory=False
)

print(df.shape)
```

```python
G = nx.Graph()
```

```python
for _, row in df.iterrows():

    founder = str(row["founder_names"])

    startup = str(row["startup_name"])

    G.add_node(
        founder,
        node_type="Founder"
    )

    G.add_node(
        startup,
        node_type="Startup"
    )

    G.add_edge(

        founder,

        startup,

        relation="FOUNDED"

    )
```

```python
for _, row in df.iterrows():

    startup = str(row["startup_name"])

    sector = str(row["sector"])

    G.add_node(
        sector,
        node_type="Sector"
    )

    G.add_edge(

        startup,

        sector,

        relation="BELONGS_TO"

    )
```

```python
for _, row in df.iterrows():

    startup = str(row["startup_name"])

    tech = str(row["technology_domain"])

    G.add_node(
        tech,
        node_type="Technology"
    )

    G.add_edge(

        startup,

        tech,

        relation="USES"

    )
```

```python
for _, row in df.iterrows():

    startup = str(row["startup_name"])

    funding = str(row["funding_stage"])

    G.add_node(
        funding,
        node_type="Funding"
    )

    G.add_edge(

        startup,

        funding,

        relation="FUNDED_AT"

    )
```

```python
for _, row in df.iterrows():

    startup = str(row["startup_name"])

    city = str(row["city"])

    G.add_node(
        city,
        node_type="City"
    )

    G.add_edge(

        startup,

        city,

        relation="LOCATED_IN"

    )
```

```python
print("Nodes :", G.number_of_nodes())

print("Edges :", G.number_of_edges())
```

```python
centrality = nx.degree_centrality(G)

top_nodes = sorted(

    centrality.items(),

    key=lambda x:x[1],

    reverse=True

)[:20]

print(top_nodes)
```

```python
startup = df.iloc[0]["startup_name"]

print("Startup:", startup)

print(

    list(
        G.neighbors(startup)
    )

)
```

```python
joblib.dump(

    G,

    "../models/startup_knowledge_graph_engine/startup_knowledge_graph.pkl"
)

print("Knowledge Graph Saved ✅")
```

```python
metadata = {

    "engine":
        "Startup Knowledge Graph",

    "nodes":
        G.number_of_nodes(),

    "edges":
        G.number_of_edges()
}

with open(

    "../models/startup_knowledge_graph_engine/metadata.json",

    "w"

) as f:

    json.dump(
        metadata,
        f,
        indent=4
    )
```

```python
import joblib
from pathlib import Path

MODELS_DIR = Path("../models")

print("="*80)
print("MODEL TYPES")
print("="*80)

checks = {

    "success_model":
        MODELS_DIR / "success_model" / "random_forest_success.pkl",

    "risk_model":
        MODELS_DIR / "risk_model" / "risk_model.pkl",

    "investor_model":
        MODELS_DIR / "investor_model" / "investor_readiness.pkl",

    "startup_health_model":
        MODELS_DIR / "startup_health_model" / "startup_health_model.pkl",

    "founder_strength_model":
        MODELS_DIR / "founder_strength_model" / "founder_strength_model.pkl",

    "funding_strength_model":
        MODELS_DIR / "funding_strength_model" / "funding_strength_model.pkl",

    "market_opportunity_model":
        MODELS_DIR / "market_opportunity_model" / "market_opportunity_model.pkl",

    "growth_potential_model":
        MODELS_DIR / "growth_potential_model" / "growth_potential_model.pkl",

    "innovation_assessment_model":
        MODELS_DIR / "innovation_assessment_model" / "innovation_assessment_model.pkl",

    "competition_analysis_model":
        MODELS_DIR / "competition_analysis_model" / "competition_analysis_model.pkl",

    "competitor_discovery_engine":
        MODELS_DIR / "competitor_discovery_engine" / "competitor_discovery_engine.pkl",

    "market_gap_detection_engine":
        MODELS_DIR / "market_gap_detection_engine" / "market_gap_detection_engine.pkl",

    "startup_clustering_engine":
        MODELS_DIR / "startup_clustering_engine" / "startup_clustering_engine.pkl",

    "startup_clustering_scaler":
        MODELS_DIR / "startup_clustering_engine" / "scaler.pkl",

    "what_if_simulator":
        MODELS_DIR / "what_if_simulator_model" / "what_if_simulator.pkl",

    "weakness_detection_engine":
        MODELS_DIR / "weakness_detection_model" / "weakness_detection_engine.pkl",

    "recommendation_engine":
        MODELS_DIR / "recommendation_model" / "full_sector_recommendation_engine.pkl",

    "knowledge_graph":
        MODELS_DIR / "startup_knowledge_graph_engine" / "startup_knowledge_graph.pkl"

}

for name, path in checks.items():

    try:

        obj = joblib.load(path)

        print(f"\n{name}")
        print(type(obj))

        if hasattr(obj, "feature_names_in_"):

            print("Features:")
            print(list(obj.feature_names_in_))

        if hasattr(obj, "classes_"):

            print("Classes:")
            print(list(obj.classes_))

        if isinstance(obj, dict):

            print("Dict Keys Sample:")
            print(list(obj.keys())[:10])

    except Exception as e:

        print(f"\n{name}")
        print("ERROR:", e)
```
