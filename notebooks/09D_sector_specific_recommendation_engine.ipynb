```python
generic_templates = {

"Founder Strength":[
"Hire experienced co-founder",
"Build advisory board",
"Increase domain expertise",
"Expand founder network",
"Improve leadership capabilities"
],

"Funding Strength":[
"Increase investor outreach",
"Improve pitch deck",
"Reduce burn rate",
"Improve unit economics",
"Prepare funding roadmap"
],

"Growth":[
"Improve customer acquisition",
"Increase retention",
"Optimize growth funnel",
"Improve conversion rate",
"Reduce churn"
],

"Market Opportunity":[
"Expand TAM",
"Target new customer segments",
"Explore adjacent markets",
"Increase market penetration",
"Expand geographically"
]

}
```

```python
def generate_v4(row):

    sector = row["sector"]
    weakness = row["weakness_1"]

    if (
        sector in sector_weakness_templates
        and
        weakness in sector_weakness_templates[sector]
    ):

        recs = sector_weakness_templates[
            sector
        ][weakness]

    else:

        recs = generic_templates[
            weakness
        ]

    return (

        recs[0],
        recs[1],
        recs[2],
        recs[3],
        recs[4],

        " → ".join(
            recs[:3]
        )

    )
```

```python
results = df.apply(
    generate_v4,
    axis=1
)

df["v4_rec_1"] = [x[0] for x in results]
df["v4_rec_2"] = [x[1] for x in results]
df["v4_rec_3"] = [x[2] for x in results]
df["v4_rec_4"] = [x[3] for x in results]
df["v4_rec_5"] = [x[4] for x in results]

df["v4_roadmap"] = [
    x[5]
    for x in results
]
```
