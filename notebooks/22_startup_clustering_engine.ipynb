```python
import pandas as pd
import numpy as np

import joblib
import json

from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(
    "../datasets/cleaned/startup_info.csv",
    low_memory=False
)

print(df.shape)
```

```python
cluster_features = [

    "startup_health_score",

    "founder_strength_score",

    "funding_strength_score",

    "market_opportunity_score",

    "growth_score",

    "innovation_score",

    "competition_score",

    "investor_readiness_score"
]
```

```python
X = df[cluster_features]
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(

    n_clusters=5,

    random_state=42,

    n_init=10
)

clusters = kmeans.fit_predict(
    X_scaled
)

df["startup_cluster"] = clusters
```

```python
print(
    df["startup_cluster"]
    .value_counts()
)
```

```python
cluster_summary = (

    df.groupby(
        "startup_cluster"
    )[cluster_features]

    .mean()

)

print(
    cluster_summary
)
```

```python
cluster_names = {

    0 : "High Growth",

    1 : "Emerging",

    2 : "Investor Ready",

    3 : "High Risk",

    4 : "Market Leaders"
}
```

```python
df["cluster_name"] = (
    df["startup_cluster"]
    .map(cluster_names)
)

print(
    df["cluster_name"]
    .value_counts()
)
```

```python
print(

    df[[
        "startup_name",

        "cluster_name"
    ]]

    .head(20)

)
```

```python
joblib.dump(

    kmeans,

    "../models/startup_clustering_engine/startup_clustering_engine.pkl"
)

joblib.dump(

    scaler,

    "../models/startup_clustering_engine/scaler.pkl"
)

print("Clustering Model Saved ✅")
```

```python
metadata = {

    "engine":

        "Startup Clustering Engine",

    "clusters":

        cluster_names,

    "features":

        cluster_features
}

with open(

    "../models/startup_clustering_engine/metadata.json",

    "w"

) as f:

    json.dump(
        metadata,
        f,
        indent=4
    )
```

```python
df.to_csv(

    "../datasets/cleaned/startup_info.csv",

    index=False
)

print(
    "Clustering Engine Complete ✅"
)
```
