```python
def detect_weaknesses(row):

    scores = {

        "Founder Strength":
            row["founder_strength_score"],

        "Funding Strength":
            row["funding_strength_score"],

        "Market Opportunity":
            row["market_opportunity_score"],

        "Growth":
            row["growth_score"]

    }

    sorted_scores = sorted(
        scores.items(),
        key=lambda x: x[1]
    )

    return (
        sorted_scores[0][0],
        sorted_scores[1][0],
        sorted_scores[2][0]
    )
```

```python
weaknesses = df.apply(
    detect_weaknesses,
    axis=1
)

df["weakness_1"] = [
    w[0]
    for w in weaknesses
]

df["weakness_2"] = [
    w[1]
    for w in weaknesses
]

df["weakness_3"] = [
    w[2]
    for w in weaknesses
]
```
