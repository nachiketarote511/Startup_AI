```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "../datasets/cleaned/startup_info.csv",
    low_memory=False
)

print(df.shape)

df.head()
```

```python
missing = (
    df.isnull()
    .sum()
    .sort_values(
        ascending=False
    )
)

missing.head(30)
```

```python
df.describe().T
```

```python
df["startup_stage"] \
.value_counts() \
.plot(kind="bar")

plt.title(
    "Startup Stage Distribution"
)

plt.show()
```

```python
df["sector"] \
.value_counts() \
.head(15) \
.plot(kind="bar")

plt.title(
    "Top Sectors"
)

plt.show()
```

```python
plt.hist(
    df["total_funding"],
    bins=50
)

plt.title(
    "Funding Distribution"
)

plt.show()
```

```python
plt.hist(
    df["startup_health_score"],
    bins=30
)

plt.title(
    "Startup Health Score"
)

plt.show()
```

```python
numeric_df = df.select_dtypes(
    include="number"
)

corr = numeric_df.corr()

corr[
    ["startup_health_score"]
].sort_values(
    by="startup_health_score",
    ascending=False
)
```

```python
top20 = df.sort_values(

    by="startup_health_score",

    ascending=False

)[[
    "startup_name",
    "sector",
    "startup_health_score"
]]

top20.head(20)
```
