```python
import pandas as pd
import joblib
import json

df = pd.read_csv(
    "../datasets/cleaned/startup_info.csv",
    low_memory=False
)

print(df.shape)
```

```python
recommendation_map = {

    "Founder Strength": [

        "Hire a technical co-founder",

        "Increase founder domain expertise",

        "Build a stronger leadership team"

    ],

    "Funding Strength": [

        "Improve investor outreach",

        "Increase revenue before fundraising",

        "Prepare investor-ready pitch deck"

    ],

    "Market Opportunity": [

        "Expand TAM by entering adjacent markets",

        "Target new customer segments",

        "Explore international expansion"

    ],

    "Growth": [

        "Improve customer acquisition",

        "Increase retention rate",

        "Optimize product-market fit"

    ]

}
```

```python
def generate_recommendations(row):

    w1 = row["weakness_1"]
    w2 = row["weakness_2"]
    w3 = row["weakness_3"]

    rec1 = recommendation_map[w1][0]
    rec2 = recommendation_map[w2][0]
    rec3 = recommendation_map[w3][0]

    return rec1, rec2, rec3
```

```python
recs = df.apply(
    generate_recommendations,
    axis=1
)

df["recommendation_1"] = [r[0] for r in recs]
df["recommendation_2"] = [r[1] for r in recs]
df["recommendation_3"] = [r[2] for r in recs]
```

```python
def get_bucket(score):

    if score <= 20:
        return "0-20"

    elif score <= 40:
        return "21-40"

    elif score <= 60:
        return "41-60"

    elif score <= 80:
        return "61-80"

    else:
        return "81-100"
```

```python
def generate_advanced_recommendation(row):

    weakness = row["weakness_1"]

    score_map = {

        "Founder Strength":
            row["founder_strength_score"],

        "Funding Strength":
            row["funding_strength_score"],

        "Market Opportunity":
            row["market_opportunity_score"],

        "Growth":
            row["growth_score"]

    }

    score = score_map[weakness]

    severity = get_severity(score)

    bucket = get_bucket(score)

    recommendations = recommendation_engine[
        weakness
    ][bucket]

    priority = (

        "High"
        if score < 40
        else "Medium"
        if score < 70
        else "Low"

    )

    roadmap = " → ".join(
        recommendations[:3]
    )

    return (
        severity,
        priority,
        "|".join(recommendations),
        roadmap
    )
```

```python
results = df.apply(
    generate_advanced_recommendation,
    axis=1
)

df["recommendation_severity"] = [
    x[0] for x in results
]

df["recommendation_priority"] = [
    x[1] for x in results
]

df["top_5_recommendations"] = [
    x[2] for x in results
]

df["improvement_roadmap"] = [
    x[3] for x in results
]
```

```python
from scipy.stats import rankdata

market_size_score = (
    rankdata(df["market_size"])
    / len(df)
) * 100

market_growth_score = (
    rankdata(df["market_growth_rate"])
    / len(df)
) * 100

tam_score = (
    rankdata(df["tam"])
    / len(df)
) * 100

sam_score = (
    rankdata(df["sam"])
    / len(df)
) * 100

som_score = (
    rankdata(df["som"])
    / len(df)
) * 100

df["market_opportunity_score"] = (

      0.30 * market_size_score

    + 0.25 * market_growth_score

    + 0.20 * tam_score

    + 0.15 * sam_score

    + 0.10 * som_score

)
```
