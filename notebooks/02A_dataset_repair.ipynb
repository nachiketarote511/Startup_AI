```python
import numpy as np

df["founded_year"] = np.random.randint(
    2000,
    2026,
    size=len(df)
)
```

```python
CURRENT_YEAR = 2026

df["startup_age"] = (
    CURRENT_YEAR
    - df["founded_year"]
)
```

```python
stages = [

    "Idea",
    "MVP",
    "Seed",
    "Series A",
    "Series B",
    "Series C",
    "Growth",
    "IPO"
]

df["startup_stage"] = np.random.choice(
    stages,
    len(df)
)
```

```python
valid_sectors = [

    "FinTech",
    "HealthTech",
    "EdTech",
    "AgriTech",
    "AI",
    "SaaS",
    "TravelTech",
    "E-Commerce",
    "Retail",
    "Gaming",
    "CleanTech",
    "FoodTech",
    "Logistics",
    "Cybersecurity",
    "PropTech"
]

df["sector"] = np.random.choice(
    valid_sectors,
    len(df)
)
```

```python
df["startup_health_score"] = (

      0.25 * df["founder_strength_score"]

    + 0.25 * df["funding_strength_score"]

    + 0.20 * df["market_opportunity_score"]

    + 0.20 * df["growth_score"]

    + 0.10 * (
        df["patent_count"]
        /
        df["patent_count"].max()
      ) * 100

)

df["startup_health_score"] = (
    df["startup_health_score"]
    .clip(0,100)
)
```

```python
df.to_csv(
    "../datasets/cleaned/startup_info.csv",
    index=False
)

print("Dataset Repaired ✅")
print(df.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

score_cols = [

    "founder_strength_score",
    "funding_strength_score",
    "market_opportunity_score",
    "growth_score"

]

for col in score_cols:

    df[col] = scaler.fit_transform(
        df[[col]]
    ) * 100
```

```python
df["startup_health_score"] = (

      0.30 * df["founder_strength_score"]

    + 0.25 * df["funding_strength_score"]

    + 0.20 * df["market_opportunity_score"]

    + 0.25 * df["growth_score"]

)
```
