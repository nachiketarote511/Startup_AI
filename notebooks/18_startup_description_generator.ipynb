```python
def generate_description(row):

    startup = str(row["startup_name"])

    sector = str(row["sector"])

    sub_sector = str(row["sub_sector"])

    tech = str(row["technology_domain"])

    business = str(row["business_model"])

    market = str(row["target_market"])

    description = f"""
{startup} is a {sector} startup operating in the {sub_sector} space.
The company leverages {tech} technology to deliver innovative solutions.
Its business model is {business} and it primarily targets {market}.
The startup focuses on scalability, customer growth, and market expansion.
"""

    return " ".join(description.split())
```

```python
df["startup_description"] = df.apply(
    generate_description,
    axis=1
)
```

```python
df = pd.read_csv(
    "../datasets/cleaned/startup_info.csv",
    low_memory=False
)

print(df.shape)
```

```python
print(
    df["startup_description"]
    .head()
)
```

```python
def description_quality(text):

    words = len(str(text).split())

    if words < 15:
        return 30

    elif words < 30:
        return 60

    elif words < 50:
        return 80

    else:
        return 100
```

```python
business_keywords = [

    "business",
    "platform",
    "service",
    "solution",
    "customer",
    "market",
    "technology"

]

def business_clarity(text):

    text = str(text).lower()

    score = 0

    for word in business_keywords:

        if word in text:
            score += 15

    return min(score,100)
```

```python
innovation_keywords = [

    "ai",
    "blockchain",
    "iot",
    "automation",
    "analytics",
    "cloud",
    "deep learning",
    "machine learning"

]

def innovation_text_score(text):

    text = str(text).lower()

    score = 0

    for word in innovation_keywords:

        if word in text:
            score += 20

    return min(score,100)
```

```python
market_keywords = [

    "customer",

    "market",

    "industry",

    "growth",

    "segment",

    "scalable"

]

def market_fit(text):

    text = str(text).lower()

    score = 0

    for word in market_keywords:

        if word in text:
            score += 15

    return min(score,100)
```

```python
investor_keywords = [

    "growth",

    "scalable",

    "technology",

    "expansion",

    "revenue",

    "innovation"

]

def investor_appeal(text):

    text = str(text).lower()

    score = 0

    for word in investor_keywords:

        if word in text:
            score += 15

    return min(score,100)
```

```python
df["description_quality_score"] = (
    df["startup_description"]
    .apply(description_quality)
)

df["business_clarity_score"] = (
    df["startup_description"]
    .apply(business_clarity)
)

df["innovation_text_score"] = (
    df["startup_description"]
    .apply(innovation_text_score)
)

df["market_fit_score"] = (
    df["startup_description"]
    .apply(market_fit)
)

df["investor_appeal_score"] = (
    df["startup_description"]
    .apply(investor_appeal)
)
```

```python
df["description_analyzer_score"] = (

      0.25 * df["description_quality_score"]

    + 0.25 * df["business_clarity_score"]

    + 0.20 * df["innovation_text_score"]

    + 0.15 * df["market_fit_score"]

    + 0.15 * df["investor_appeal_score"]

)
```

```python
print(
    df[[
        "description_quality_score",
        "business_clarity_score",
        "innovation_text_score",
        "market_fit_score",
        "investor_appeal_score",
        "description_analyzer_score"
    ]].describe()
)
```

```python
import json

metadata = {

    "analyzer":"Startup Description Analyzer",

    "output_scores":[

        "description_quality_score",

        "business_clarity_score",

        "innovation_text_score",

        "market_fit_score",

        "investor_appeal_score",

        "description_analyzer_score"
    ]

}

with open(

    "../models/startup_description_analyzer/metadata.json",

    "w"

) as f:

    json.dump(
        metadata,
        f,
        indent=4
    )
```
